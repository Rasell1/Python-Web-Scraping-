```python
from bs4 import BeautifulSoup
import  requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_video_game_companies_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('table')[1]
```

```python
table = soup.find('table', class_ ="wikitable sortable")
```

```python
titles_tag = table.tr.find_all('th')
table_title = [title.text.strip() for title in titles_tag]
table_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_title)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
data = []
```

```python
for row in column_data:
    row_data = row.find_all('td')
    data.append([data.text.strip() for data in row_data])
    
```

```python
print(data)
```

```python
data2 = pd.DataFrame(data,columns=table_title)
data2
```

```python
data2.to_csv(r"D:\Rasel\Data analytics\Python\project\video_game_companies.csv", index=False)
print("CSV file saved as 'video_game_companies.csv'")
```
